```python
"""
Basic Statistical Operations Program

This script allows the user to perform simple operations on a list of numbers.
The available operations are:

1. Sum      - Adds all the numbers together
2. Average  - Calculates the mean value
3. Maximum  - Finds the largest number
4. Minimum  - Finds the smallest number

How it works:
- The user chooses an operation by entering a number (1–4).
- Then they input a list of numbers separated by spaces.
- The program converts the input into a list of floats and performs the chosen operation.
- Results are printed directly to the console.
- Error handling is included to catch invalid inputs (like non-numeric values or wrong menu choices).
"""

def sum(numbers):
    total = 0
    for num in numbers:
        total += num
    return total
def average(numbers):
    return sum(numbers) / len(numbers)
def maximum(numbers):
    maximum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
    return maximum
def minimum(numbers):
    minimum = numbers[0]
    for num in numbers:
        if num < minimum:
            minimum = num
    return minimum
print("Choose an operation:")
print("1: Sum | 2: Average | 3: Maximum | 4: Minimum")
try:
    choice = int(input("Enter your choice (1-4): "))
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(x) for x in user_input.split()]
    if choice == 1:
              print("Sum:", sum(numbers))
    elif choice == 2:
                print("Average:", average(numbers))
    elif choice == 3:
                print("Maximum:", maximum(numbers))
    elif choice == 4:
                print("Minimum:", minimum(numbers))
    else:
        print("Error: Enter a valid number (1-4).")
except ValueError:
    print("Error: Invalid input. Please enter numbers correctly.")
```

```python
"""
Sublist Extraction Program

This script allows the user to filter numbers from a list based on a given
range. The user provides:

- A list of numbers separated by spaces
- A starting number (b)
- An ending number (c)

The function `get_sublist(a, b, c)` then checks each element of the list `a`
and appends it to a new list `d` if the element falls within the specified
range. The resulting sublist is printed.

Example:
Input list: 5 10 15 20 25
Start: 10
End: 20
Output: [10, 15]
"""

def get_sublist(a, b, c):
  d=[]
  i=0
  while i<=c:
    if a[i] in range(b,c):
      d.append(a[i])
    i=i+1
  print(d)
no=input("enter numbers with space \t")
a=[int(x) for x in no.split()]
b=int(input("enter the starting no for the list \t"))
c=int(input("enter the ending no for the list \t"))
get_sublist(a,b,c)
```

```python
"""
Flatten Nested List Program

This script takes a nested list (a list of lists) and flattens it into
a single list containing all the elements.

How it works:
- The function `flatten(lst)` loops through each sublist inside `lst`.
- Then it loops through each element of that sublist and appends it
  to a new list `d`.
- Finally, the flattened list is printed.

Example:
Input: [[1, 2], [3, 4], [5]]
Output: [1, 2, 3, 4, 5]
"""
def flatten(lst):
  d=[]
  for b in lst:
    for c in b:
      d.append(c)
  print(d)
lst=[[1, 2], [3, 4], [5]]
flatten(lst)
```

**3. Sum of All Elements in a Nested List:**

```python
"""
Sum Nested List Program

This script calculates the total sum of all numbers inside a nested list.
It can handle lists that contain other lists inside them (up to two levels deep).

How it works:
- The function `sum_nested(lst)` loops through each element of the list.
- If the element is itself a list, it loops through that sublist.
- If the sublist contains another list, it loops through that too.
- Every number found is added to the running total.
- Finally, the sum is printed.

Example:
Input: [[1, 2], [3, [4, 5]], 6]
Output: 21
"""
def sum_nested(lst):
  sum = 0
  for x in lst:
    if isinstance(x,list):
      for y in x:
        if isinstance(y,list):
          for z in y:
            sum = sum + z
        else:
          sum = sum + y
    else:
      sum = sum + x
  print(sum)
lst = [[1, 2], [3, [4, 5]], 6]
sum_nested(lst)
```

**4. Remove Specific Element from a Nested List:**

```python
"""
Remove Specific Value from Nested List Program

This script removes all occurrences of a given number from a nested list.
It works recursively, meaning it can handle lists inside lists at any depth.

How it works:
- The function `remove(lst, n)` loops through each element in the list.
- If the element is itself a list, the function calls itself again to
  process that sublist.
- If the element is not a list, it checks whether the value matches `n`.
  If it does not match, the element is kept.
- The final list is returned with all instances of `n` removed.

Example:
Input: [[1, 2], [3, [4, 5]], 6]
Remove: 2
Output: [[1], [3, [4, 5]], 6]
"""
def remove(lst,n):
  b=[]
  for x in lst:
    if isinstance(x, list):
      b.append(remove(x, n))
    else:
      if x != n:
        b.append(x)
  return(b)
a = [[1, 2], [3, [4, 5]], 6]
print(remove(a,2))
```
