```
#importing the image from the local device
from google.colab import files
uploaded = files.upload()
```

```
#importing all the necessary modules
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```
image_name = list(uploaded.keys())[0]
img = Image.open(image_name).convert("RGB")
img_np = np.array(img)
# row column channel
top_left_100 = img_np[0:100, 0:100, :]
plt.imshow(top_left_100)
plt.axis("off")
plt.title("this is 100 X 100")
plt.show()
```

```
# Extract individual channels
R = img_np.copy()
G = img_np.copy()
B = img_np.copy()
R[:, :, 1] = 0
R[:, :, 2] = 0
G[:, :, 0] = 0
G[:, :, 2] = 0
B[:, :, 0] = 0
B[:, :, 1] = 0
plt.imshow(R)
plt.title("R")
plt.axis('off')
plt.show()
plt.imshow(G)
plt.title("G")
plt.axis('off')
plt.show()
plt.imshow(B)
plt.title("B")
plt.axis('off')
plt.show()
```

```
# Make a copy of the original image
modified_img = img_np.copy()
# Ensure dtype is uint8
modified_img = modified_img.astype(np.uint8)
# Set top-left 100x100 pixels to 210 for all channels
modified_img[0:100, 0:100, :] = 210
# Display
plt.imshow(modified_img)
plt.axis('off')
plt.title("Modified Image (Top-left 100x100 = 210)")
plt.show()
```
